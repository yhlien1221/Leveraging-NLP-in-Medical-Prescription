```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time

# Set up WebDriver with manually installed ChromeDriver
service = Service("/Users/yhlien/.wdm/drivers/chromedriver/mac64/133.0.6943.53/chromedriver-mac-arm64/chromedriver")  # Update with actual path
driver = webdriver.Chrome(service=service)

# Start WebDriver
driver = webdriver.Chrome(service=service, options=options)

# Open the Cleveland Clinic Diseases page
url = "https://my.clevelandclinic.org/health/diseases"
driver.get(url)
time.sleep(3)  # Wait for the page to fully load

# Find all letter buttons dynamically
letters = driver.find_elements(By.CSS_SELECTOR, "a.js-enabled")

# Dictionary to store diseases categorized by letter
all_diseases = {}

# Loop through each letter button
for letter in letters:
    letter_text = letter.text.strip()
    
    if letter_text in "ABCDEFGHIJKLMNOPQRSTUVWXYZ":  # Ensure it's a valid letter
        print(f"Clicking letter: {letter_text}")

        # Click the letter button using JavaScript
        driver.execute_script("arguments[0].click();", letter)
        time.sleep(3)  # Wait for the diseases list to load

        # Extract disease names and links
        diseases = driver.find_elements(By.CSS_SELECTOR, "div.results__single.results__sliding-item > a")

        disease_list = []
        for disease in diseases:
            disease_name = disease.find_element(By.TAG_NAME, "h4").text.strip()  # Extract disease name
            disease_url = disease.get_attribute("href")  # Extract disease link

            if disease_name and disease_url:
                disease_list.append((disease_name, disease_url))

        # Store in dictionary
        all_diseases[letter_text] = disease_list

# Close browser
driver.quit()

# Print extracted data
for letter, diseases in all_diseases.items():
    print(f"\nDiseases under '{letter}':")
    for name, link in diseases:
        print(f"{name}: {link}")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

# For explicit wait
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Set up Chrome WebDriver
service = Service(ChromeDriverManager().install())
options = Options()
options.headless = True  # Run in headless mode (no UI)

# Start WebDriver
driver = webdriver.Chrome(service=service, options=options)

# Open the Rabies disease page
url = "https://my.clevelandclinic.org/health/diseases/13848-rabies"
driver.get(url)
time.sleep(3)  # Let page load

# -----------------------------------------------------------------------------
# 1) Extract Disease Name (e.g., <h1>)
# -----------------------------------------------------------------------------
try:
    disease_name = driver.find_element(By.TAG_NAME, "h1").text.strip()
except:
    disease_name = "N/A"

# -----------------------------------------------------------------------------
# 2) Extract Overview
# -----------------------------------------------------------------------------
try:
    overview_element = driver.find_element(By.CSS_SELECTOR, "p[data-identity='paragraph-element']")
    overview_text = overview_element.text.strip()
except:
    overview_text = "N/A"

# -----------------------------------------------------------------------------
# 3) Extract ONLY the Prodromal symptoms (li items) under that section
# -----------------------------------------------------------------------------
try:
    # Wait until the h4 with the text "Prodromal symptoms of rabies" is present
    prodromal_heading = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, "//h4[contains(text(), 'Prodromal symptoms of rabies')]"))
    )

    # Find the <ul> that immediately follows that heading
    symptoms_ul = prodromal_heading.find_element(By.XPATH, "./following-sibling::ul[@data-identity='unordered-list']")

    # Get all the <li> elements under that <ul>
    symptom_elements = symptoms_ul.find_elements(By.TAG_NAME, "li")

    # Build a clean list of symptom texts
    symptoms_list = [li.text.strip() for li in symptom_elements if li.text.strip()]

    # Join them into a single string (adjust separator as needed)
    symptoms_text = "; ".join(symptoms_list)
except:
    symptoms_text = "N/A"

# -----------------------------------------------------------------------------
# 4) Extract Diagnosis items
# -----------------------------------------------------------------------------
diagnosis_list = []
try:
    diagnosis_items = driver.find_elements(
        By.CSS_SELECTOR,
        "li.mb-rem12px.pl-rem12px.text-gray-800.text-rem19px.leading-rem34px"
    )

    for item in diagnosis_items:
        try:
            strong_text = item.find_element(By.TAG_NAME, "strong").text.strip()
            if strong_text.endswith("."):
                strong_text = strong_text[:-1].strip()
            diagnosis_list.append(strong_text)
        except:
            pass

    diagnosis_text = ",".join(diagnosis_list) if diagnosis_list else "N/A"
except:
    diagnosis_text = "N/A"

# -----------------------------------------------------------------------------
# 5) Extract Other Sections (Treatment, Drugs Used)
# -----------------------------------------------------------------------------
sections = ["Treatment", "Drugs Used"]
disease_data = {
    "Disease Name": disease_name,
    "Overview": overview_text,
    "Symptoms": symptoms_text,
    "Diagnosis": diagnosis_text
}

for section in sections:
    try:
        # Find the section header
        section_element = driver.find_element(By.XPATH, f"//h2[contains(text(), '{section}')]")

        # Get the text content of the next sibling <p> tag (description under the section)
        try:
            section_text = section_element.find_element(By.XPATH, "./following-sibling::p").text.strip()
        except:
            section_text = "N/A"

    except:
        section_text = "N/A"

    # --------------------------------------------------------------
    # Append medication names ONLY if this is the "Treatment" section
    # --------------------------------------------------------------
    if section == "Treatment":
        try:
            # Find all medication names inside the treatment list
            medication_elements = driver.find_elements(
                By.XPATH,
                "//ul[@data-identity='unordered-list']//li"
                "//a[@data-identity='link' and contains(@href,'health/drugs')]/strong"
            )

            # Extract text and clean
            meds = [med.text.strip().rstrip('.') for med in medication_elements]

            if meds:
                # Ensure treatment text is included
                if section_text and section_text != "N/A":
                    section_text += "\n\nMedications Mentioned: " + ", ".join(meds)
                else:
                    # If no treatment text is found, fallback to just medications
                    section_text = ", ".join(meds)
        except:
            pass

    disease_data[section] = section_text

# Close browser
driver.quit()

# -----------------------------------------------------------------------------
# 6) Convert to DataFrame
# -----------------------------------------------------------------------------
df = pd.DataFrame([disease_data])

# Save the extracted data to CSV
df.to_csv("cleveland_clinic_disease_information.csv", index=False)

# Display DataFrame
df
```
